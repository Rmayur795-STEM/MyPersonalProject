### 2. Initialize the Generative Model

Before you can make any API calls, you need to initialize the Generative Model. We'll use a `gemini-2.5-flash-preview-04-17` model, which is good for quick, high-quality responses.

```python
import google.generativeai as genai

# Initialize the Gemini API model using the updated package
gemini_model = genai.GenerativeModel('gemma-3-1b-it')
#gemini_model = genai.GenerativeModel('gemini-flash-latest')
```

### 3. Make an API Call to Generate Content

Now you can make API calls to generate text. For your chatbot, you would dynamically insert student queries here. For this template, let's try a generic college-related query.

```python
# Example query for your chatbot
student_query = "What are some key things to include in a college application essay?"

# Generate content using the model
response = gemini_model.generate_content(student_query)

# Print the AI's response
print("AI Chatbot Response:\n")
print(response.text)
```

```python
print('Available models that support generateContent:')
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```
